```python
df = df.withColumnRenamed('Item_Outlet_Sales', 'SaLeS')
display(df)
```

```python
from pyspark.sql.functions import when, col

df = df.withColumn(
    "MRP_Category",
    when(col("Item_MRP") > 200, "Premium")
    .otherwise("Non-Premium")
)

display(df)
```

```python
from pyspark.sql.functions import *

df = df.withColumn(
    "Category",
    when(
        col("Item_MRP") > 200,
        "Premium"
    ).otherwise("Normal")
)

display(df)
```

```python
df = df.withColumn(
    "Special",
    when(
        (col("Item_MRP") > 200) &
        (col("SaLeS") > 5000),
        "Yes"
    )
    .otherwise("No")
)
display(df)
```

```python
df.groupBy(
    "Outlet_Type"
).sum(
    "SaLeS"
).show()
```

```python
from pyspark.sql.functions import sum

df.groupBy("Outlet_Type").pivot("Item_Type").agg(sum("SaLeS")).display()
```

```python
from pyspark.sql.functions import sum


df.groupBy(
    "Outlet_Type",
    "Outlet_Size"
).pivot(
    "Item_Fat_Content"
).agg(
    sum("SaLeS")
).display()
```

```python
print(df.columns)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col,sum
window_spec = Window.partitionBy(
    "Outlet_Type"
)

df = df.withColumn(
    "Total_Sales",
    sum("SaLeS")
    .over(window_spec)
)

display(df)
```
